```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('C:/Users/acanto/Desktop/heart.csv')
```

```python
pd.crosstab(data['age'],data['target']).plot(kind="bar",figsize=(20,6))
plt.title("Heart Disease Frequency In Different Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(data['sex'], hue=data['target'])
plt.title("Gender based Rates")
plt.xlabel("Male:1,Female:0")
plt.ylabel("Number of Counts")
plt.show()
```

```python
X = data.drop('target', axis=1)
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
svc_model = SVC()
regression_model = LogisticRegression()
```

```python
param_grid = {'C':[0.01,0.1,1.0,10,100,1000,10000,100000],
             'gamma':[1,0.1,0.01,0.001,0.0001,0.00001,0.000001,0.0000001]}
grid = GridSearchCV(svc_model,param_grid,verbose=2)
grid.fit(X_train, y_train)
```

```python
regression_model.fit(X_train, y_train)
```

```python
# Accuracy on training data of SVC model
X_train_pred_svc = grid.predict(X_train)
svc_training_data_accuracy = accuracy_score(y_train, X_train_pred_svc)
print(svc_training_data_accuracy)
```

```python
# Accuracy on training data of SVC model
X_test_pred_svc = grid.predict(X_test)
svc_testing_data_accuracy = accuracy_score(y_test, X_test_pred_svc)
print(svc_testing_data_accuracy)
```

```python
# Calculate predictions for both models on training data
X_train_pred_regression = regression_model.predict(X_train)
X_train_pred_svc = grid.predict(X_train)

mse_regression_train = mean_squared_error(y_train, X_train_pred_regression)
mse_svc_train = mean_squared_error(y_train, X_train_pred_svc)

models = ['Regression', 'SVC']
mse_values_test = [mse_regression_train, mse_svc_train]
print(mse_values_test)

plt.bar(models, mse_values_test, color=['blue', 'green'])
plt.xlabel('Models')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Comparison of Training MSE between Regression and SVC Models')
plt.show()
```

```python
# Calculate predictions for both models on testing data
X_test_pred_svc = grid.predict(X_test)
X_test_pred_regression = regression_model.predict(X_test)

mse_regression_test = mean_squared_error(y_test, X_test_pred_regression)
mse_svc_test = mean_squared_error(y_test, X_test_pred_svc)

mse_test_values = [mse_regression_test, mse_svc_test]
print(mse_test_values)

plt.bar(models, mse_test_values, color=['blue', 'green'])
plt.xlabel('Models')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Comparison of Testing MSE between SVC and Regression Models')
plt.show()
```
